```python
import pandas as pd
```

```python
shark = pd.read_csv('attacks.csv', encoding = "ISO-8859-1")
```

```python
def process_columns(columns):
    res = []
    for col in columns:
        coli = col.strip().lower()
        res.append(coli)
    return res
```

```python
shark.columns = process_columns(shark.columns)
```

```python
shark.set_index('Case Number')
```

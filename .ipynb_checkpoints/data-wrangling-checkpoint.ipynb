```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
sharkattacks = "./attacks.csv"
```

```python
dataf = pd.read_csv(sharkattacks)
```

```python
dataf.info()
```

```python
c_todrop = ['Case Number.1','Case Number.2','Name','pdf','href','href formula','original order','Injury','Time','Investigator or Source','Unnamed: 22','Unnamed: 23']
dataf.drop(columns= c_todrop, inplace=True) #dropping columns that are irrelevant
```

```python
dataf.info()
```

```python
dataf
```

```python
dataf.dropna(how='all',inplace=True)
```

```python
dataf.info()
```

```python
dataf.dropna(thresh=4,inplace=True)
```

```python
dataf.columns 
```

```python
dataf.rename(columns={'Case Number': 'case_number','Date':'date','Year':'year','Type':'type','Country':'country', 'Are':'area', 'Location':'location','Activity':'activity','Sex ':'gender', 'Age':'age', 'Fatal (Y/N)':'fatal_Y/N', 'Species ':'species'},inplace=True)
```

```python
dataf['type'].value_counts() # will delete column as it seems vast majority is unprovoked. 
```

```python
dataf.drop(columns=['type'],inplace=True)
```

```python
dataf
```

```python
dataf.drop_duplicates(inplace=True)
```

```python
dataf.fillna('Unknown')
```

```python
dataf.reset_index(inplace=True)
```

```python
under1960 =[x for x in dataf['year'] if x <1960]
```

```python
dataf.sort_values('year',inplace=True)
dataf
```

```python
dataf.set_index('year',inplace=True)
dataf
```

```python
dataf.drop(index= under1960,inplace=True) # dropping rows under year 1960
```

```python
dataf
```

```python
dataf['species'].value_counts() # will drop as it seems not a lot of knowledge on shark species
```

```python
dataf.drop(columns=['species'],inplace=True)
```

```python
dataf['case_number'].replace(to_replace=r"\.[a-zA-Z]",value='',regex=True,inplace=True)
```

```python
dataf
```

```python
dataf.reset_index(inplace=True)
```

```python
dataf.tail()
```

```python
dataf.drop(index=[4034],inplace=True)
```

```python
dataf.drop(columns=['year','index'],inplace=True)
```

```python
#replacing values on case number to convert to correct date

dataf['case_number'].replace(to_replace=r"00",value='01',regex=True,inplace=True)
dataf['case_number'].replace(to_replace=r"02.29",value='02.28',regex=True,inplace=True)
dataf['case_number'].replace(to_replace=r"\D",value='.',regex=True,inplace=True)
dataf['case_number'].replace(to_replace=r"13\d",value='13',regex=True,inplace=True)
```

```python
dataf['attack_date'] = pd.to_datetime(dataf['case_number'])
```

```python
dataf.set_index('attack_date', inplace=True)
```

```python
dataf
```

```python
dataf.drop(columns=['case_number','date'],inplace=True)
```

```python
dataf
```

```python
dataf.fillna(value='Unknown',inplace=True)
```

```python
dataf['gender'].value_counts()
```

```python
dataf['gender'].replace(to_replace=' ',value='',regex=True).replace(to_replace='lli',value='Unknown',regex=True, inplace=True)
```

```python
dataf['age'].value_counts()
```

```python
dataf.drop(columns=['age'],inplace=True) #dropping age because seems irrelevant as 1/3rd are unknown
```

```python
dataf['fatal_Y/N'].value_counts()
```

```python
dataf['fatal_Y/N'].replace(to_replace='\w{2,}',value='Unknown',regex=True,inplace=True)
dataf['fatal_Y/N'].replace(to_replace='\s*',value='',regex=True,inplace=True)
dataf['fatal_Y/N'].replace(to_replace='M',value='N',regex=True,inplace=True)
```

```python
dataf['fatal_Y/N'].value_counts()
```

```python
dataf.sort_values('attack_date')
```

```python
dataf.reset_index(inplace=True)
dataf
```

```python
dataf.rename(columns={'index':'Case Number','country':'Country','location':'Location','activity':'Activity','gender':'Gender','fatal_Y/N':'Fatal_Y/N'},inplace=True)
```

```python
dataf.set_index(['Case Number'],inplace=True)
```

```python
dataf
```

```python
dataf.to_csv('clean-shark-attacks.csv',index=False)
```

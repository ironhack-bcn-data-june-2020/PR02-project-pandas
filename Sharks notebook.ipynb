Aplicaremos reglas para aquellas observaciones para los que haya algo de información y aproximaremos el mes. Por ejemplo, si el evento ocurrió en verano, asignaremos el mes de julio.

```python
# Aplicamos reglas aproximando estaciones del año, early, late y 2 casos que aparecen -Ap- (entendemos que es abril)

for i,c in df[df['Month'].isna()].iterrows():
    if str(c['Date']).lower().find('summer') != -1:
        df.loc[i, 'Month'] = 'Jul'
    elif str(c['Date']).lower().find('winter') != -1:
        df.loc[i, 'Month'] = 'Jan'
    elif str(c['Date']).lower().find('fall') != -1:
        df.loc[i, 'Month'] = 'Oct'
    elif str(c['Date']).lower().find('late') != -1:
        df.loc[i, 'Month'] = 'Dec'
    elif str(c['Date']).lower().find('early') != -1:
        df.loc[i, 'Month'] = 'Jan'
    elif str(c['Date']).lower().find('-Ap-') != -1:
        df.loc[i, 'Month'] = 'Apr'
```

Como hemos empezado a extraer los valores en texto (Jan, Feb, etc.), tendremos todos en texto y después los pasaremos a entero. Iteraremos en el campo 'Date' con regex para buscar toda cifra de 2 dígitos tras 4 dígitos, de forma que responda al patrón YYYY.MM y pasaremos ese valor a un diccionario para que saque el mes correspondiente con texto.

```python
# Creamos un diccionario para pasar a texto lo que extraigamos en la siguiente iteración

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
numonths = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
dicmonths = dict(zip(numonths, months))
dicmonths['01']
```

```python
# Iteramos para que nos encuentre toda cifra de 2 dígitos entre puntos tras otra de 4 dígitos (YYYY.MM)

# Le pasamos el valor del diccionario creado anteriormente

for i,c in df[df['Month'].isna()].iterrows():
    try:
        a = re.findall('\d{4}\.(\d{2})', c['Date'])[0]
        df.loc[i, 'Month'] = dicmonths[a]
    except:
        pass
```

```python
# Cambiamos los 2 valores restantes para los que hay info de mes manualmente consultándo el índice y haciendo .loc

df[(df['Date'] == '12-30-1980') | (df['Date'] == '06-26-1890')]
```

### Day

En el caso de la info sobre el día, la he sacado de 'Case Number', ya que la info en la columna 'Date' estaba demasiado confusa.

```python
# Ya que el día es complicado de sacar de la columna 'Date', lo sacamos de 'Case Number' con regex

for i,c in df.iterrows():
    try:
        x = re.findall('\d{4}\.\d{2}\.(\d{2})', c['Case Number'])[0]
        if x == '00':
            df.loc[i, 'Day'] = 1            # Coge el día 1 de cada mes por defecto si no hay datos
        elif x.startswith('0'):
            df.loc[i, 'Day'] = int(x[-1])   # Coge el 2º valor si empieza por 0 y lo pasa a int
        else:
            df.loc[i, 'Day'] = int(x)
    except:
        df.loc[i, 'Day'] = 1
        pass
```

En este apartado se pueden ver las operaciones finales en el dataframe antes de exportarlo a csv.

```python
df.head()
```

```python
# Guardamos el índice original con reset_index

df.reset_index(inplace=True)
```

```python
# Lo renombramos a 'Original order inv'

df.rename(columns = {'index': 'Index'}, inplace=True)
```

```python
# Ordenamos la columna 'Datetime' antes de fijarla como índice

df.sort_values(['Datetime'], ascending = False, inplace=True)
```

```python
# Miramos cuáles son las observaciones previas al año 1700, porque darán problemas para fijar la fecha como índice

for i,c in df.iterrows():
    print(i, c['Datetime'])
```

```python
# Eliminamos dichas observaciones

df.drop(df.index[6279:], inplace=True)
```

```python
# Fijamos la fecha como índice para que sea time series

df.set_index('Datetime', inplace=True)
```

```python
df.head(3)
```

```python
# Las columnas 'Year' y 'Month' ya están incluidas en la fecha 'Datetime', las eliminamos

df.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
```

```python
# Creamos una nueva columna 'Date Info' concatenando la info de fecha y hora

for i,c in df.iterrows():
    a = ''
    if c['Date'] != 'Unknown':
        a = a + str(c['Date']) + ', '
    elif c['Time'] != 'Unknown':
        a = a + str(c['Location']) + ', '
    df.loc[i, 'Date Info'] = a[:-2]
```

```python
# Ahora ya podemos eliminar también las columnas 'Date' y 'Time'

df.drop(['Date', 'Time'], axis=1, inplace=True)
```

```python
# Actualizamos el dataframe para orderar las columnas

df = df[['Name', 'Sex', 'Age', 'Date Info', 'Location Info', 'Country', 'ISO code','Continent', 'Main activity',
     'Activity', 'Type', 'Injury', 'Fatal (Y/N)', 'Main Species','Info Species', 
     'Investigator or Source', 'Index', 'Case Number', 'pdf', 'href formula']]
```

Aplicaremos reglas para aquellas observaciones para los que haya algo de informaciÃ³n y aproximaremos el mes. Por ejemplo, si el evento ocurriÃ³ en verano, asignaremos el mes de julio.

```python
# Aplicamos reglas aproximando estaciones del aÃ±o, early, late y 2 casos que aparecen -Ap- (entendemos que es abril)

for i,c in df[df['Month'].isna()].iterrows():
    if str(c['Date']).lower().find('summer') != -1:
        df.loc[i, 'Month'] = 'Jul'
    elif str(c['Date']).lower().find('winter') != -1:
        df.loc[i, 'Month'] = 'Jan'
    elif str(c['Date']).lower().find('fall') != -1:
        df.loc[i, 'Month'] = 'Oct'
    elif str(c['Date']).lower().find('late') != -1:
        df.loc[i, 'Month'] = 'Dec'
    elif str(c['Date']).lower().find('early') != -1:
        df.loc[i, 'Month'] = 'Jan'
    elif str(c['Date']).lower().find('-Ap-') != -1:
        df.loc[i, 'Month'] = 'Apr'
```

Como hemos empezado a extraer los valores en texto (Jan, Feb, etc.), tendremos todos en texto y despuÃ©s los pasaremos a entero. Iteraremos en el campo 'Date' con regex para buscar toda cifra de 2 dÃ­gitos tras 4 dÃ­gitos, de forma que responda al patrÃ³n YYYY.MM y pasaremos ese valor a un diccionario para que saque el mes correspondiente con texto.

```python
# Creamos un diccionario para pasar a texto lo que extraigamos en la siguiente iteraciÃ³n

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
numonths = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
dicmonths = dict(zip(numonths, months))
dicmonths['01']
```

```python
# Iteramos para que nos encuentre toda cifra de 2 dÃ­gitos entre puntos tras otra de 4 dÃ­gitos (YYYY.MM)

# Le pasamos el valor del diccionario creado anteriormente

for i,c in df[df['Month'].isna()].iterrows():
    try:
        a = re.findall('\d{4}\.(\d{2})', c['Date'])[0]
        df.loc[i, 'Month'] = dicmonths[a]
    except:
        pass
```

```python
# Cambiamos los 2 valores restantes para los que hay info de mes manualmente consultÃ¡ndo el Ã­ndice y haciendo .loc

df[(df['Date'] == '12-30-1980') | (df['Date'] == '06-26-1890')]
```

### Day

En el caso de la info sobre el dÃ­a, la he sacado de 'Case Number', ya que la info en la columna 'Date' estaba demasiado confusa.

```python
# Ya que el dÃ­a es complicado de sacar de la columna 'Date', lo sacamos de 'Case Number' con regex

for i,c in df.iterrows():
    try:
        x = re.findall('\d{4}\.\d{2}\.(\d{2})', c['Case Number'])[0]
        if x == '00':
            df.loc[i, 'Day'] = 1            # Coge el dÃ­a 1 de cada mes por defecto si no hay datos
        elif x.startswith('0'):
            df.loc[i, 'Day'] = int(x[-1])   # Coge el 2Âº valor si empieza por 0 y lo pasa a int
        else:
            df.loc[i, 'Day'] = int(x)
    except:
        df.loc[i, 'Day'] = 1
        pass
```

# Estructura final del dataframe

En este apartado se pueden ver las operaciones finales en el dataframe antes de exportarlo a csv.

```python
df.head()
```

```python
# Guardamos el Ã­ndice original con reset_index

df.reset_index(inplace=True)
```

```python
# Lo renombramos a 'Original order inv'

df.rename(columns = {'index': 'Index'}, inplace=True)
```

```python
# Ordenamos la columna 'Datetime' antes de fijarla como Ã­ndice

df.sort_values(['Datetime'], ascending = False, inplace=True)
```

```python
# Miramos cuÃ¡les son las observaciones previas al aÃ±o 1700, porque darÃ¡n problemas para fijar la fecha como Ã­ndice

for i,c in df.iterrows():
    print(i, c['Datetime'])
```

```python
# Eliminamos dichas observaciones

df.drop(df.index[6279:], inplace=True)
```

```python
# Fijamos la fecha como Ã­ndice para que sea time series

df.set_index('Datetime', inplace=True)
```

```python
df.head(3)
```

```python
# Las columnas 'Year' y 'Month' ya estÃ¡n incluidas en la fecha 'Datetime', las eliminamos

df.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
```

```python
# Creamos una nueva columna 'Date Info' concatenando la info de fecha y hora

for i,c in df.iterrows():
    a = ''
    if c['Date'] != 'Unknown':
        a = a + str(c['Date']) + ', '
    elif c['Time'] != 'Unknown':
        a = a + str(c['Location']) + ', '
    df.loc[i, 'Date Info'] = a[:-2]
```

```python
# Ahora ya podemos eliminar tambiÃ©n las columnas 'Date' y 'Time'

df.drop(['Date', 'Time'], axis=1, inplace=True)
```

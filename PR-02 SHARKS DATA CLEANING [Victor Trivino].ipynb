**Which is the average probability from 2000 per sex, that if a shark attacks you, will be fatal?**

# Pre-analysis

Importing the corresponding libraries

```python
import numpy as np
import pandas as pd
import re
```

## We upload the dataset to Jupyter Notebook

```python
sharks = pd.read_csv("C:/Users/victo/Desktop/IronHack/PR02-project-pandas/attacks.csv", encoding = "ISO-8859-1")
sharks
```

```python
sharks.columns
```

## We create a new dataframe 'sharks1' with the columns only needed to our analysis

```python
sharks1 = sharks[['Date', 'Year', 'Name', 'Sex ', 'Fatal (Y/N)']] 
sharks1
```

## We drop NA and duplicate cases for removing errors in the dataset


```python
sharks1.dropna(axis=0, how='any', inplace=True)
sharks1
```

```python
sharks1 = sharks1.drop_duplicates()
sharks1
```

# Columns analysis and cleaning

## Analysis of 'Date' and 'Year' observations

### We convert the date to a string

```python
sharks1['Date'] = sharks1['Date'].astype(str)
```

### We define a function to convert the year in the right format for our analysis. We use regex and find all for finding the right results

```python
def find_year(date):
    try:
        year = [int(i) for i in list(re.findall(r'.*([1-3][0-9]{3})', date))]
        return int(np.mean(year))
    except:
        return 0

sharks1['Year'] = sharks1['Date'].apply(find_year)
sharks1
```

### We check that the funcion has worked properly by using value_counts

```python
sharks1['Year'].value_counts()
```

### We filter the results from 2000

```python
sharks1= sharks1[sharks1['Year']>=2000]    
sharks1
```

### We check that the filter is right by using unique

```python
sharks1['Year'].unique()
```

## Analysis of 'Name' and 'Sex ' observations

### Analysis of Name. We observe that male and female are not names, they are part of Sex column

```python
sharks1['Name'].value_counts()
```

### Analysis of Sex . We observe inconsistencies because there are some female names which are M in the column sex

```python
sharks1[sharks1.Name=='female'].sort_values(by='Sex ', ascending=False)
```

#### For solving the issue we create a function and we use assign to add it as a new column

```python
def newSex(x):
    if x['Name'] == 'female':
        return 'F'
    elif x['Name'] == 'male':
        return 'M'
    elif (x['Name'] != 'female') and (x['Sex '] == 'F'):
        return 'F'
    elif (x['Name'] != 'male') and (x['Sex '] == 'M'):
        return 'M'
    else:
        return x['Sex ']

sharks1 = sharks1.assign(RightSex=sharks1.apply(newSex, axis=1))
sharks1
```

#### We check that the issue has been solved by using the same syntax as previously

```python
sharks1[sharks1.Name=='female'].sort_values(by='Sex ', ascending=False)
```

#### Analysis of RightSex data

```python
sharks1['RightSex'].unique()
```

```python
sharks1['RightSex'].value_counts()
```

##### To correct the mistakes, we drop 'lli' case and we replace 'M ' by 'M' 

```python
sharks1.drop(sharks1[(sharks1.RightSex=='lli')].index, inplace=True)
sharks1.replace(to_replace ='M ', value='M', inplace = True)
```

##### We check that both cases have disappeared

```python
sharks1['RightSex'].value_counts()
```

## Analysis of Fatal observations

```python
sharks1['Fatal (Y/N)'].value_counts()
```

### We rename the column to 'Fatal' for not having issues later on

```python
sharks1.rename(columns={'Fatal (Y/N)': 'Fatal'}, inplace = True)
```

### We drop the invalid cases 'UNKNOWN', 'M', '2017'

```python
sharks1.drop(sharks1[(sharks1.Fatal== 'UNKNOWN') | (sharks1.Fatal=='M') | (sharks1.Fatal=='2017')].index, inplace=True)
```

### We check that those cases have disappeared

```python
sharks1['Fatal'].value_counts()
```

# Pivot table and summary

## We create the pivot table into a new dataframe


```python
sharks2 = pd.pivot_table(sharks1,index='Year', columns=('RightSex', 'Fatal'), values='Date', 
                         aggfunc=[len], fill_value=0, margins=True)
sharks2
```

## We convert the pivot table into percentages

```python
sharks2 = sharks2.div( sharks2.iloc[:,-1], axis=0 )
sharks2
```
